```python
import numpy as np

# Define the vectors
a = np.array([0, 1, 0, 9, 1])
b = np.array([3, 0, 0, 9, 0])

# Compute the dot product of a and b
dot_product = np.dot(a, b)

# Compute the norm (magnitude) of each vector
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

# Compute the cosine similarity
cosine_similarity = dot_product / (norm_a * norm_b)

# Compute the cosine similarity loss (1 - cosine similarity)
cosine_similarity_loss = 1 - cosine_similarity

# Output the result
print(f"Cosine Similarity: {cosine_similarity}")
print(f"Cosine Similarity Loss: {cosine_similarity_loss}")
```

```python
# تعریف بردارها
a = np.array([0, 1, 0, 9, 1])
b = np.array([3, 0, 0, 9, 0])

# نرمال‌سازی برای نمایش جهت
a_norm = a / np.linalg.norm(a)
b_norm = b / np.linalg.norm(b)

# انتخاب دو بعد برای نمایش (مثلاً مؤلفه‌های index 1 و 3)
a_proj = np.array([a_norm[1], a_norm[3]])
b_proj = np.array([b_norm[1], b_norm[3]])

# رسم نمودار
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, a_proj[0], a_proj[1], angles='xy', scale_units='xy', scale=1, color='red', label='بردار a')
plt.quiver(0, 0, b_proj[0], b_proj[1], angles='xy', scale_units='xy', scale=1, color='blue', label='بردار b')
plt.xlim(-0.2, 1)
plt.ylim(-0.2, 1)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title("نمایش شباهت دو بردار")
plt.xlabel("بعد اول (عنصر 1)")
plt.ylabel("بعد دوم (عنصر 3)")
plt.show()
```

```python
# نرمال‌سازی برای نمایش جهت
a_norm = a / np.linalg.norm(a)
b_norm = b / np.linalg.norm(b)

# انتخاب سه مؤلفه اصلی برای نمایش در فضای سه‌بعدی (مثلاً 0، 1 و 3)
a_3d = np.array([a_norm[0], a_norm[1], a_norm[3]])
b_3d = np.array([b_norm[0], b_norm[1], b_norm[3]])

# ایجاد شکل سه‌بعدی
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# رسم بردارها از مبدأ
ax.quiver(0, 0, 0, a_3d[0], a_3d[1], a_3d[2], color='r', label='بردار a')
ax.quiver(0, 0, 0, b_3d[0], b_3d[1], b_3d[2], color='b', label='بردار b')

# تنظیم محدوده نمایش
ax.set_xlim([-0.2, 1])
ax.set_ylim([-0.2, 1])
ax.set_zlim([0, 1])

# افزودن برچسب‌ها و عنوان
ax.set_xlabel('بعد 0')
ax.set_ylabel('بعد 1')
ax.set_zlabel('بعد 3')
ax.set_title('نمایش سه‌بعدی بردارهای a و b')
ax.legend()
plt.tight_layout()
plt.show()
```
